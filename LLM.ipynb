```python
# تثبيت المكتبات
!pip install -q transformers sentencepiece ipywidgets

# استيراد الأدوات
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import ipywidgets as widgets
from IPython.display import display, clear_output

# تحميل النموذج والتوكن
tokenizer = AutoTokenizer.from_pretrained("microsoft/GODEL-v1_1-large-seq2seq")
model = AutoModelForSeq2SeqLM.from_pretrained("microsoft/GODEL-v1_1-large-seq2seq")

# إعداد الشخصية الافتراضية للنموذج (تعاطفية)
persona = "I am a caring and supportive AI assistant."

# إنشاء واجهة المستخدم
text_input = widgets.Textarea(
    value='',
    placeholder='اكتب شعورك أو كلامك هنا...',
    description='أنت:',
    layout=widgets.Layout(width='100%', height='80px')
)

send_button = widgets.Button(
    description='إرسال',
    button_style='success'
)

output = widgets.Output()

# دالة الرد من GODEL
def godel_reply(user_input):
    instruction = f"[CONTEXT] {user_input} [ENDOFTURN] [USER] {user_input} [ENDOFTURN] [SYSTEM]"
    input_text = f"{persona} {instruction}"

    inputs = tokenizer(input_text, return_tensors="pt", padding=True)
    outputs = model.generate(**inputs, max_new_tokens=100)
    response = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return response

# عند الضغط على زر الإرسال
def handle_send(b):
    user_msg = text_input.value.strip()
    output.clear_output()

    if user_msg == "":
        with output:
            print("⚠️ الرجاء كتابة نص للرد عليه.")
        return

    response = godel_reply(user_msg)

    with output:
        print(f"🧑‍💻 أنت: {user_msg}")
        print(f"🤖 الذكاء العاطفي: {response}")

    text_input.value = ''

send_button.on_click(handle_send)

# عرض الواجهة
display(text_input, send_button, output)
```
